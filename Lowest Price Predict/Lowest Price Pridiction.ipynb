```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Preprocessed_Trian_data.csv')
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df.head()
```

```python
unscaled_inputs = df.drop('Low_Cap_Price',axis=1)
targets = df.iloc[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_inputs = scaler.fit_transform(unscaled_inputs)
```

```python
scaled_inputs
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(scaled_inputs,targets,test_size=20,random_state=12)
```

```python
y_train.shape
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
linReg = LinearRegression()
linReg.fit(x_train,y_train)
```

```python
linReg.score(x_train,y_train)
```

```python
y_pred = linReg.predict(x_test)
```

```python
prediction = pd.DataFrame()
prediction['Low_Cap_Price Given'] = y_test
prediction['Lowest Price Predicted'] = y_pred
prediction.head()
```

```python
# Testing the model
test_data = pd.read_csv('Preprocessed_Test_data.csv')
test_data.drop('Unnamed: 0',axis=1,inplace=True)
test_data.head()
```
